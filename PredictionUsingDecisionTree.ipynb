```python
# importing required libraries
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
```

```python
# Loading iris dataset
data = load_iris()
```

```python
data.data.shape
```

```python
print('Classes to Predict: ', data.target_names)
```

```python
x = data.data
y = data.target
```

```python
display(x.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Defining test and train dataset
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = 50, test_size = 0.25)
```

```python
# Defining the decision tree classifier and fitting test and train dataset
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
clf.fit(x_train, y_train)
```

```python
# Predicting output on test dataset
y_pred = clf.predict(x_test)
```

```python
# Finding the accuracy score
from sklearn.metrics import accuracy_score
print('Accuracy score on train data: ', accuracy_score(y_true=y_train, y_pred=clf.predict(x_train)))
print('Accuracy score on test data: ', accuracy_score(y_true=y_test, y_pred=y_pred))
```

```python
# Visualizing Decision Tree using graph viz
from six import StringIO
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus

# Visualize the graph
dot_data = StringIO()
export_graphviz(clf, out_file=dot_data, feature_names=data.feature_names,  
                filled=True, rounded=True,
                special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```
